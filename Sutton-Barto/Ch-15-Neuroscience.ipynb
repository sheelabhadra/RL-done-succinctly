# Chapter 15: Neuroscience

* A neuron consists of 
    * a cell body
    * dendrites: recieve inputs from other neurons
    * a single axon: fiber that carries the neuron's output to other neurons
    <img src="Figures/NeuronStructures.jpg" style="width: 550px;"/>

* **Action potentials:** the output of a neuron that consists of a sequence of electrical pulses that travel along the axon. Action potentials are also called spikes and a neuron is said to fire when it generates a spike.
* **Firing rate:** the average number of spikes per some unit of time
* **Axonal arbor:** the branching structure of a neuron's axon
* **Background activity:** firing rate when the neuron doesn't appear to be driven by synaptic input
* **Phasic activity:** spiking activity usually generated by synaptic input
* **Tonic activity:** activity that varies slowly and often in a graded manner
* **Efficacy:** The strength or effectiveness by which the neurotransmitter released at a synapse influences the postsynaptic neuron is the synapse's efficacy
* **Synaptic plasticity:** The ability of synaptic efficacies to change. It is one of the primary mechanisms responsible for learning.
